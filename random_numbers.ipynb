```python
# Import the NumPy module.
import numpy as np

# Import the random module.
import random 

# Import timeit. (To test how long a piece of code will take to run)
import timeit
```

```python
# This give you a random number between 0 and 1 giving a large decimal number
random.random()
```

```python
# This gives you a random integer within a range.
random.randint(-10,10)
```

```python
# This is similar to the random.randint() but the potential random values that can be generated by random.randrange ()
# are limited by the step argument. {it takes steps from the lower boundary to the upper boundary}
random.randrange(-9,9,2)
```

```python
# This gives you a number with multiple decimal places within the given range
random.uniform(-10,10)
```

```python
# This While statement allows us to get 10 random numbers. The addition is to get decimals
x = 1
latitudes = []
while x < 11:
    random_lat = random.randint(-90, 89) + random.random()
    latitudes.append(random_lat)
    x += 1
latitudes
```

```python
# The np allows you get get as many random numbers as you wish (this one has 50) from the range you selected.
np.random.uniform(-90,90,50)
```

```python
%timeit np.random.uniform(-90.000, 90.000, size=1500)
```

```python
def latitudes(size):
    latitudes = []
    x = 0
    while x < (size):
        random_lat = random.randint(-90, 90) + random.random()
        latitudes.append(random_lat)
        x += 1
    return latitudes
# Call the function with 1500. 
%timeit latitudes(1500)
```
